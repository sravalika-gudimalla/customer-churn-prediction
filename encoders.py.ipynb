```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle

# Load your dataset
df = pd.read_csv(r"C:\Users\sravalika\Downloads\Customer Churn.csv")

# Drop customerID if exists
if "customerID" in df.columns:
    df = df.drop("customerID", axis=1)

# List of all categorical columns your model uses
cat_cols = [
    'gender','Partner','Dependents','PhoneService','MultipleLines',
    'InternetService','OnlineSecurity','OnlineBackup','DeviceProtection',
    'TechSupport','StreamingTV','StreamingMovies',
    'Contract','PaperlessBilling','PaymentMethod'
]

# Create a LabelEncoder for each categorical column
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Save the encoders dictionary
with open("encoders.pkl", "wb") as f:
    pickle.dump(label_encoders, f)

print("✔️ encoders.pkl UPDATED SUCCESSFULLY with all categorical features!")
print("Columns encoded:", list(label_encoders.keys()))
```
