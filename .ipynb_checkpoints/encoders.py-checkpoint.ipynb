```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle

# Load your Kaggle dataset
df = pd.read_csv(r"C:\Users\sravalika\Downloads\Customer Churn.csv")

# Remove customerID column if present
if "customerID" in df.columns:
    df = df.drop("customerID", axis=1)

# Columns exactly as used in your Streamlit app
cat_cols = ['Gender','Partner','Dependents','PhoneService','MultipleLines','InternetService',
            'OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']

# Dictionary to store encoders
label_encoders = {}

# Fit LabelEncoder for each categorical column
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Save encoders.pkl using pickle
with open("encoders.pkl", "wb") as f:
    pickle.dump(label_encoders, f)

print("✔️ encoders.pkl created successfully!")
print("Columns encoded:", list(label_encoders.keys()))
```
